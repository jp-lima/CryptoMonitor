```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep
from io import StringIO
```

```python
driver=webdriver.Chrome()

driver.get("https://www.coingecko.com/pt?items=300")
```

```python
lista_todas_criptomoedas = []
for n in range (0,6):
    df = pd.read_html(io= StringIO(driver.page_source))
    
    
    dados_coletados = pd.concat(df)
    
    dados_coletados.drop(['Unnamed: 0','#', 'Unnamed: 3', 'Últimos 7 dias','1 h','24 h','7 d','30 d', 'Capitalização de Mercado / FDV'], axis=1, inplace=True)
    
    dados_coletados.drop ([0], axis=0, inplace=True)
    
    
    
    lista_todas_criptomoedas.append(dados_coletados)
    
    driver.find_element(By.CSS_SELECTOR, 'a[aria-label="next"]').click()
    
    

df_todas_criptomoedas = pd.concat(lista_todas_criptomoedas)
driver.close()
```

```python

display(df_todas_criptomoedas)

df_todas_criptomoedas.to_csv('criptomoedas.csv', index=False)
```
